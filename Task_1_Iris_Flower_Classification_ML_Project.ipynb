**Model Building**

```python
X = iris_data.drop(['class'], axis=1)
y = iris_data['class']
print(f'X shape: {X.shape} | y shape: {y.shape} ')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=1)
```

```python
model = []
model.append(('SVC', SVC(gamma='auto')))
```

```python
model = SVC(gamma='auto')
model.fit(X_train, y_train)
prediction = model.predict(X_test)
```

```python
print(f'Test Accuracy: {accuracy_score(y_test, prediction)}')
print(f'Classification Report: \n {classification_report(y_test, prediction)}')
```

**Result**:

The accuracy on the test set is 93% this model made the prediction for 93% of the irises in the given dataset.We can expect 93% accuracy in predictikng the new iris species.
